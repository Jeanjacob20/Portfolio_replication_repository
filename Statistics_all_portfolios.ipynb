```python
# Cell 1: Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuration de l'affichage
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', 50)
pd.set_option('display.float_format', lambda x: f'{x:.4%}' if abs(x) < 1 else f'{x:.4f}')

print("✓ Imports réussis")
```

```python
# Cell 2: Chargement du DataFrame principal
df_results = pd.read_csv('results/tables/evaluation_all_models_all_portfolios.csv', encoding='utf-8-sig')

print(f"Shape du DataFrame: {df_results.shape}")
print(f"\nColonnes: {list(df_results.columns)}")
print(f"\nNombre de portefeuilles: {len(df_results)}")
print("\nPremières lignes:")
df_results.head()
```

```python
# Cell 3: Affichage complet du DataFrame
print("="*100)
print("TABLEAU COMPLET - ÉVALUATION DE TOUS LES MODÈLES")
print("="*100)
print(df_results.to_string(index=False))
```

```python
# Cell 4: Statistiques descriptives sur les RMSE
print("="*100)
print("STATISTIQUES DESCRIPTIVES - RMSE OUT-OF-SAMPLE")
print("="*100)

models_rmse = {
    'Linear_Simple': 'RMSE_Linear_Simple',
    'Linear_WF': 'RMSE_Linear_WF',
    'Ridge': 'RMSE_Ridge',
    'Genetic': 'RMSE_Genetic',
    'MeanVar': 'RMSE_MeanVar'
}

stats_rmse_list = []

for model_name, rmse_col in models_rmse.items():
    if rmse_col in df_results.columns:
        rmse_values = df_results[rmse_col].dropna()
        if len(rmse_values) > 0:
            stats_rmse_list.append({
                'Modèle': model_name,
                'Moyenne': rmse_values.mean(),
                'Médiane': rmse_values.median(),
                'Min': rmse_values.min(),
                'Max': rmse_values.max(),
                'Écart-type': rmse_values.std(),
                'Q1 (25%)': rmse_values.quantile(0.25),
                'Q3 (75%)': rmse_values.quantile(0.75),
                'N_Valides': len(rmse_values)
            })

df_stats_rmse = pd.DataFrame(stats_rmse_list)
print(df_stats_rmse.to_string(index=False))
```

```python
# Cell 5: Statistiques descriptives sur la stabilité (MAD)
print("="*100)
print("STATISTIQUES DESCRIPTIVES - STABILITÉ (MAD)")
print("="*100)

models_stability = {
    'Linear_WF': 'Stabilite_MAD_Linear_WF',
    'Ridge': 'Stabilite_MAD_Ridge',
    'Genetic': 'Stabilite_MAD_Genetic',
    'MeanVar': 'Stabilite_MAD_MeanVar'
}

stats_stability_list = []

for model_name, mad_col in models_stability.items():
    if mad_col in df_results.columns:
        mad_values = df_results[mad_col].dropna()
        if len(mad_values) > 0:
            stats_stability_list.append({
                'Modèle': model_name,
                'MAD Moyenne': mad_values.mean(),
                'MAD Médiane': mad_values.median(),
                'MAD Min': mad_values.min(),
                'MAD Max': mad_values.max(),
                'MAD Écart-type': mad_values.std(),
                'N_Valides': len(mad_values)
            })

df_stats_stability = pd.DataFrame(stats_stability_list)
print(df_stats_stability.to_string(index=False))
```

```python
# Cell 6: Statistiques descriptives sur le Turnover
print("="*100)
print("STATISTIQUES DESCRIPTIVES - TURNOVER")
print("="*100)

models_turnover = {
    'Linear_WF': 'Stabilite_Turnover_Linear_WF',
    'Ridge': 'Stabilite_Turnover_Ridge',
    'Genetic': 'Stabilite_Turnover_Genetic',
    'MeanVar': 'Stabilite_Turnover_MeanVar'
}

stats_turnover_list = []

for model_name, turnover_col in models_turnover.items():
    if turnover_col in df_results.columns:
        turnover_values = df_results[turnover_col].dropna()
        if len(turnover_values) > 0:
            stats_turnover_list.append({
                'Modèle': model_name,
                'Turnover Moyen': turnover_values.mean(),
                'Turnover Médian': turnover_values.median(),
                'Turnover Min': turnover_values.min(),
                'Turnover Max': turnover_values.max(),
                'Turnover Écart-type': turnover_values.std(),
                'N_Valides': len(turnover_values)
            })

df_stats_turnover = pd.DataFrame(stats_turnover_list)
print(df_stats_turnover.to_string(index=False))
```

```python
# Cell 8: Boxplot comparatif des RMSE
fig, ax = plt.subplots(figsize=(14, 8))

rmse_data = []
rmse_labels = []

for model_name, rmse_col in models_rmse.items():
    if rmse_col in df_results.columns:
        rmse_values = df_results[rmse_col].dropna()
        if len(rmse_values) > 0:
            rmse_data.append(rmse_values)
            rmse_labels.append(model_name)

bp = ax.boxplot(rmse_data, labels=rmse_labels, patch_artist=True)

# Colorier les boxplots
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']
for patch, color in zip(bp['boxes'], colors[:len(bp['boxes'])]):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)

ax.set_ylabel('RMSE', fontsize=12)
ax.set_title('Comparaison des RMSE par modèle\n(Boxplot)', fontsize=14, fontweight='bold')
ax.grid(True, alpha=0.3, axis='y')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('results/figures/boxplot_rmse_comparaison.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Cell 3: Calcul du turnover moyen par méthode
print("="*100)
print("TURNOVER MOYEN PAR MÉTHODE")
print("="*100)

models_turnover = {
    'Linear_WF': 'Stabilite_Turnover_Linear_WF',
    'Ridge': 'Stabilite_Turnover_Ridge',
    'Genetic': 'Stabilite_Turnover_Genetic',
    'MeanVar': 'Stabilite_Turnover_MeanVar'
}

turnover_stats = []

for model_name, turnover_col in models_turnover.items():
    if turnover_col in df_results.columns:
        turnover_values = df_results[turnover_col].dropna()
        if len(turnover_values) > 0:
            turnover_stats.append({
                'Méthode': model_name,
                'Turnover_Moyen': turnover_values.mean(),
                'Turnover_Médian': turnover_values.median(),
                'Turnover_Min': turnover_values.min(),
                'Turnover_Max': turnover_values.max(),
                'Turnover_Écart-type': turnover_values.std(),
                'N_Valides': len(turnover_values)
            })

df_turnover_stats = pd.DataFrame(turnover_stats)
print("\nStatistiques du Turnover par méthode:")
print(df_turnover_stats.to_string(index=False))
```

```python
# Cell 4: Visualisation - Bar plot du turnover moyen par méthode
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Graphique 1: Bar plot du turnover moyen
ax1 = axes[0]
if len(df_turnover_stats) > 0:
    methods = df_turnover_stats['Méthode'].values
    turnover_means = df_turnover_stats['Turnover_Moyen'].values
    
    bars = ax1.bar(methods, turnover_means, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'][:len(methods)], 
                   alpha=0.7, edgecolor='black')
    
    # Ajouter les valeurs sur les barres
    for bar, value in zip(bars, turnover_means):
        height = bar.get_height()
        ax1.text(bar.get_x() + bar.get_width()/2., height,
                f'{value:.4f}',
                ha='center', va='bottom', fontsize=10, fontweight='bold')
    
    ax1.set_ylabel('Turnover Moyen', fontsize=12)
    ax1.set_title('Turnover Moyen par Méthode', fontsize=14, fontweight='bold')
    ax1.grid(True, alpha=0.3, axis='y')
    plt.setp(ax1.xaxis.get_majorticklabels(), rotation=45, ha='right')

# Graphique 2: Boxplot du turnover par méthode
ax2 = axes[1]
turnover_data = []
turnover_labels = []

for model_name, turnover_col in models_turnover.items():
    if turnover_col in df_results.columns:
        turnover_values = df_results[turnover_col].dropna()
        if len(turnover_values) > 0:
            turnover_data.append(turnover_values)
            turnover_labels.append(model_name)

if len(turnover_data) > 0:
    bp = ax2.boxplot(turnover_data, labels=turnover_labels, patch_artist=True)
    
    # Colorier les boxplots
    colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']
    for patch, color in zip(bp['boxes'], colors[:len(bp['boxes'])]):
        patch.set_facecolor(color)
        patch.set_alpha(0.7)
    
    ax2.set_ylabel('Turnover', fontsize=12)
    ax2.set_title('Distribution du Turnover par Méthode', fontsize=14, fontweight='bold')
    ax2.grid(True, alpha=0.3, axis='y')
    plt.setp(ax2.xaxis.get_majorticklabels(), rotation=45, ha='right')

plt.suptitle('Analyse du Turnover par Méthode', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('results/figures/turnover_par_methode.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Cell 10: Tableau comparatif des meilleurs modèles par portefeuille
print("="*100)
print("MEILLEUR MODÈLE PAR PORTEFEUILLE (plus bas RMSE)")
print("="*100)

# Trouver le meilleur modèle pour chaque portefeuille
best_models = []

for idx, row in df_results.iterrows():
    portfolio_name = row['Portefeuille_Thematique']
    
    # Comparer tous les RMSE
    rmse_values = {}
    for model_name, rmse_col in models_rmse.items():
        if rmse_col in row and pd.notna(row[rmse_col]):
            rmse_values[model_name] = row[rmse_col]
    
    if rmse_values:
        best_model = min(rmse_values, key=rmse_values.get)
        best_rmse = rmse_values[best_model]
        
        best_models.append({
            'Portefeuille_Thematique': portfolio_name,
            'Meilleur_Modèle': best_model,
            'RMSE_Meilleur': best_rmse,
            'Tous_RMSE': rmse_values
        })

df_best = pd.DataFrame(best_models)
print(df_best[['Portefeuille_Thematique', 'Meilleur_Modèle', 'RMSE_Meilleur']].to_string(index=False))

# Compter combien de fois chaque modèle est le meilleur
print("\n" + "="*100)
print("FRÉQUENCE - Nombre de fois où chaque modèle est le meilleur")
print("="*100)
best_model_counts = df_best['Meilleur_Modèle'].value_counts()
print(best_model_counts.to_string())
```
